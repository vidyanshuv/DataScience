# Grouping Data

### By Column Index

```python
address = r"C:\Users\vvishal\Documents\DataScienceTutorials\Data\mtcars.csv"

# Read the CSV file
cars = pd.read_csv(address)

# Create columns for our dataframe
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

# Print 5 entries of the dataframe 
cars.head()
```

```python
car_group = cars.groupby(cars['cyl'])
car_group.mean()
```

```python
car_group = cars.groupby(cars['am'])
car_group.mean()
```

```python
car_subset = cars[['mpg', 'disp', 'hp', 'wt']]
car_subset
```
