```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
DF_obj = DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj_1 = DataFrame(np.arange(15).reshape(5, 3))
DF_obj_1
```

# Concatenation 

concat() method concatinates the dataframes by rows unless provided an axix value e.g. x.concat([df1, df2], axis = 1)

```python
pd.concat([DF_obj, DF_obj_1], axis = 1)
```

```python
pd.concat([DF_obj, DF_obj_1])
```

# Transformation

### drop()
drop() method drops specific rows from a dataframe, same drops a column when provided with an axis

```python
DF_obj.drop([0, 2])
```

```python
DF_obj.drop([0, 2], axis = 1)
```

### Adding Data

```python
series_obj = Series(np.arange(6))
series_obj.name = "Added Variable"
series_obj
```

```python
var_added = DataFrame.join(DF_obj, series_obj)
var_added
```

```python
added_datatable = var_added.append(var_added, ignore_index = False)
added_datatable
```

### Sorting Data

```python
df_sorted = DF_obj.sort_values(by=(5), ascending = False)
df_sorted
```
